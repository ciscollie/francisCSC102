```python
#  Write a Python program to calculate the roots of a cubic equation
from scipy.special import cbrt
# User inserts coefficients of x in cubic equation
m = float(input("Insert the coefficient of x^3: "))
n = float(input("Insert the coefficient of x^2: "))
o = float(input("Insert the coefficient of x: "))
p = float(input("Insert the constant term: "))
# Make the cubic equation monic
# Divide through by cofficient of x^3
a1 = n/m
a2 = o/m
a3 = p/m
# Formula for solving cubic equations
Q = ((3*a2)-((a1)**2))/9
R = ((9*a1*a2)-(27*a3)-(2*((a1)**3)))/54
b = (((Q)**3) + ((R)**2))
#Conditions for is b is greater or less than 0
def wahala(b):
    if b < 0:
        return -(-R+((((Q)**3) + ((R)**2))**(1/2)))**(1/3)
    else:
        return cbrt(R+((((Q)**3) + ((R)**2))**(1/2)))
def problem(b):
    if b < 0:
        return -(-R-((((Q)**3) + ((R)**2))**(1/2)))**(1/3)
    else:
        return cbrt(R-((((Q)**3) + ((R)**2))**(1/2)))
S = wahala(b)
T = problem(b)

#Formula for finding the roots of the equation
x1 = ((S + T)-((1/3)*a1))
y = ((1/2)*(complex(0,1))*(3**(1/2))*(S - T))
z2 = ((-1/2)*(S + T)-((1/3)*a1))
x2 = ((-1/2)*(S + T)-((1/3)*a1) + y)
z3 = ((-1/2)*(S + T)-((1/3)*a1))
x3 = ((-1/2)*(S + T)-((1/3)*a1) - y)
# The roots of the cubic equation
print("The roots of the cubic equation are: ")
print("x1: ", x1)
if y == 0:
    print("x2: ",z2)
else:
    print("x2: ",x2)
if y == 0:
    print("x3: ",z3)
else:
    print("x3: ",x3)
```

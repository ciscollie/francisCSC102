```python
import math
from abc import ABC, abstractmethod
class Shape(ABC):
    @abstractmethod
    def calculate_area(self):
        pass
class Triangle(Shape):
    def __init__(self, thebase, theheight):
        self.base = thebase
        self.height = theheight
    def calculate_area(self):
        return int(0.5 * self.base * self.height)
class Rectangle(Shape):
    def __init__(self, thelength, thebreadth):
        self.length = thelength
        self.breadth = thebreadth
    def calculate_area(self):
        return self.length * self.breadth
class Square(Shape):
    def __init__(self, theside):
        self.side = theside
    def calculate_area(self):
        return int(self.side)**2
class Circle(Shape):
    def __init__(self, theradius):
        self.radius = theradius
    def calculate_area(self):
        return math.pi * int((self.radius))**2
shapes = ['triangle', 'rectangle', 'square', 'circle']
enter = str(input("What shape do you want to find the area: "))
enter = enter.lower()
def shape():
    if enter == 'rectangle':
        a = int(input("Insert the length: "))
        b = int(input("Insert the breadth: "))
        rec = Rectangle(a,b)
        print("Area of the rectangle is:", rec.calculate_area())
    elif enter == 'circle':
        c = int(input("Insert the radius: "))
        cir = Circle(c)
        print("Area of the circle is:", cir.calculate_area())
    elif enter == 'square':
        d = int(input('Insert the side of the square: '))
        squ = Square(d)
        print("Area of the square is:", squ.calculate_area())
    elif enter == 'triangle':
        e = int(input('Insert the base of the triangle: '))
        f =int(input('Input the height of the triangle: '))
        tri = Triangle(e,f)
        print("Area of the triangle is: ", tri.calculate_area())
shape()

while enter not in shapes:
    for x in shapes:
        print(x)
    print("Cannot find area of ", enter, "please select shapes above")
    enter = input("What shape do you want to find the area: ")
    shape()

    
    
    

    
  




```

```python
from abc import ABC, abstractmethod
class Bird(ABC):
    def defecate(self):
        return 'All birds defecate'
    @abstractmethod
    def Move(self):
        pass

class Hen(Bird):
    def Move(self):
        return 'The Hen moves by Waddling'
class Ostrich(Bird):
    def Move(self):
        return 'The Ostrich runs and can run really fast at about 73Km/hr'
bird1 = Hen()
bird2 = Ostrich()
print(bird1.defecate())
print(bird1.Move())
print(bird2.defecate())
print(bird2.Move())

    
```
